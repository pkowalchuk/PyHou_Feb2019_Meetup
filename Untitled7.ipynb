```python
import matplotlib.pyplot as plt
```

```python
image=plt.imread('Equalized_Hawkes_Bay_NZ.jpg')
print(type(image))

plt.title('original image')
plt.axis('off')
plt.imshow(image)
plt.show()
```

```python
pixels=image.flatten()

plt.subplot(2,1,2)
plt.xlim(0,255)
plt.title('hystogram')
plt.hist(pixels,bins=64,range=(0,256),color='red',alpha=0.4)
plt.show()
```

```python
print(pixels.shape)
print(type(pixels[1]))

import numpy as np

pixels_int = pixels.astype(np.int32)
print(type(pixels_int[1]))
```

```python
%%R -i pixels_int

hist(pixels_int)
```

```python
%%R -i pixels_int -o pixels_r

print(typeof(pixels_int))
print(dim(pixels_int))
pixels_r<-ifelse(pixels_int<200,pixels_int,0)
print(head(pixels_int))
print(head(pixels_r))
print(min(pixels_r))
print(max(pixels_r))
```

```python
type(pixels_r)
```

```python
pixels_from_r=np.array(pixels_r)
```

```python
type(pixels_from_r)
```

```python
image_from_r=pixels_from_r.reshape(683,1024,3)
```

```python
plt.title('image from r')
plt.axis('off')
plt.imshow(image_from_r,cmap='gray')
plt.show()
```
